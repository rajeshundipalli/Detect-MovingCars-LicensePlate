## Prepocessing

### Create a function to convert image and display the image

```python
def func(image,gray=False,size=(100,100)):
    temp=image
    if gray==False:
        temp=cv2.cvtColor(temp,cv2.COLOR_BGR2RGB)
        plt.imshow(temp,cmap='gray')
        plt.show()
        
```

### Display the image

```python
img=cv2.imread("CarLicensePlate3.jpg")
func(img)
```

## Create function for classifier

```python
a = cv2.CascadeClassifier("haarcascade_russian_plate_number.xml")#Read the file
```

## Detect Car License Plate

```python

def detect_number(img):
    temp=img
    gray=cv2.cvtColor(temp,cv2.COLOR_BGR2GRAY)
    number=a.detectMultiScale(img,1.2)
    print(str(len(number)))
    for (x,y,w,h) in number:
        #roi_gray=gray[y:y+h,x:x+w]
        #roi_color= img[y:y+h,x:x+h]
        cv2.rectangle(temp,(x,y),(x+w,y+h),(0,255,0),3)
    func(temp)
    
```

## Display Results

```python
img=cv2.imread("CarLicensePlate3.jpg")
detect_number(img)
```

## Alternate

```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numberPlate_cascade = "haarcascade_russian_plate_number.xml"
detector = cv2.CascadeClassifier(numberPlate_cascade)
```

```python
img = cv2.imread('CarLicensePlate3.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
number=detector.detectMultiScale(img_gray,1.2)
#detector.detectMultiScale() takes an image as input and returns a list containing the location of detected objects in the input image.
number
```

```python
for (x,y,w,h) in number:
    
    # draw bounding box
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    # Crop the numberplate
    plateROI = img_gray[y:y+h,x:x+w]
    cv2.imshow("Numberplate", plateROI)
    
# Show the final output
cv2.imshow('Output', img)
# wait until  any key is pressed
cv2.waitKey(0)
```
